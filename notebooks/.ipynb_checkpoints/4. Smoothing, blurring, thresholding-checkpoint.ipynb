Bluring
- Average
- Guassian
- Median
- Bilateral

### Average
- define k*k window, where k is an odd number
- the pixel at the center of the matrix(anchor) is set to the average of the rest (k-1) pixels
- as the size of the kernel increases, the more blurred our image will become.

````python
blurred = np.hstack([
    cv2.blur(image, (3,3)),    # (3,3) is the suze of the kernel
    cv2.blur(image, (5,5)),
    cv2.blur(image, (7,7))
])
cv2_to_PIL(blurred)
````

### Gaussian
- Instead of using the simple mean, use weighted mean
- ```cv2.GaussianBlur``` paramters:
    - image
    - kernel size
    - $ \sigma $: Standard deviation in the x direction
        - 0 = automatically compute them based on our kernel size.

````python
blurred = np.hstack([
    cv2.GaussianBlur(image,(3,3),0),
    cv2.GaussianBlur(image,(5,5),0),
    cv2.GaussianBlur(image,(7,7),0)
])
cv2_to_PIL(blurred)
````

### Median
- Most effective to remove salt and pepper noise, because each central pixel is always replaced with a pixel intensity that exists in the image.
- replace the central pixel with the median of the neighborhood.

````python
blurred = np.hstack([
    cv2.medianBlur(image,3),
    cv2.medianBlur(image,5),
    cv2.medianBlur(image,7)
])
cv2_to_PIL(blurred)
````

### Bilateral
- While blurring , we tend to lose the edges in the image.
- to reduce noise while still maintaining edges, we can use bilateral blurring.
- This is accomplished by introducing two Gaussian distributions.
- The first Gaussian function only considers spatial neighbors, that is, pixels that appear close together in the (x, y) coordinate space of the image. 
- The second Gaussian then models the pixel intensity of the neighborhood, ensuring that only pixels with similar intensity are included in the actual computation of the blur.
- It is considerably slower than its averaging, Gaussian, and median blurring
- ```cv2.bilateralFilter``` parameters
    - image
    - diamter of the pixel neighborhood
    - color: A larger value for color means that more colors in the neighborhood will be considered when computing the blur.
    - space: A larger value of space means that pixels farther out from the central pixel will influence the blurring calculation, provided that their colors are similar enough.

````python
blurred = np.hstack([
    cv2.bilateralFilter(image, 5, 21, 21),
    cv2.bilateralFilter(image, 7, 31, 31),
    cv2.bilateralFilter(image, 9, 41, 41)
])
cv2_to_PIL(blurred)
````

### Thresholding
- to segment the foreground from the background
- convert a grayscale image to a binary image, where the pixels are either 0 or 255.
- A simple thresholding example would be selecting a pixel value p, and then setting all pixel intensities less than p to zero, and all pixel values greater than p to 255.

#### Simple thresholding
- Requires human intervention to set the __Threshold__
- value < threshold = 0
- value > threshold = 255
- Inverse thresholding can also be applied.

````python
image = cv2.imread('../data/coins.png')
cv2_to_PIL(image)
````

````python
# To remove the high frequency edges
blurred_img = cv2.GaussianBlur(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), (5,5), 0)
cv2_to_PIL(blurred_img)
````

```cv2.threshold``` parameters:
- grayscale image
- threshold, T
- Max value applied during threshold
- Thresholding method<br><br>
```cv2.threshold``` return values:
- Threshold, T
- Thresholded image

````python
(T, threshold) = cv2.threshold(blurred_img, 155, 255, cv2.THRESH_BINARY)
cv2_to_PIL(threshold)
````

````python
#Inverse blur
(T, thresholdInv) = cv2.threshold(blurred_img, 155, 255, cv2.THRESH_BINARY_INV)
cv2_to_PIL(threshold)
````

````python
#Make the RoIs as coins areas 
cv2_to_PIL(cv2.bitwise_and(image,image,mask=thresholdInv))
````

#### Adaptive Thresholding
- considers small neighbors of pixels and then finds an optimal threshold value T for each neighbor.
```cv2.adaptiveThreshold``` parameters:
- image
- maximum value
- method to compute the threshold for the current neighborhood of pixels
    - cv2.ADAPTIVE_THRESH_MEAN_C: Treat the mean of the neighborhood as T
    - cv2.ADAPTIVE_THRESH_GAUSSIAN_C: Treat the weighted mean of the neighborhood as T
- thresholding method
- neighborhood size
- C = integer that is subtracted from the mean, allowing us to fine-tune our thresholding

````python
adp_threshold = cv2.adaptiveThreshold(blurred_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 4)
cv2_to_PIL(adp_threshold)
````

````python
adp_gaussian_threshold = cv2.adaptiveThreshold(blurred_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 3)
cv2_to_PIL(adp_gaussian_threshold)
````
