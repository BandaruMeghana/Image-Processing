This is because, openCV reads the images in BGR order rather than RGB.Therefore convert BGR to RGB before displaying.

````python
plt.axis('off')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
````

````python
#Dsiplaying using PIL
def cv2_to_PIL(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    pil_image = Image.fromarray(image)
    return pil_image
````

````python
cv2_to_PIL(image)
````

### Accessing & Manipulating pixels

````python
#Accessing the pixel
(b,g,r) = image[0,0]
#We have white in the (0,0) pixel. RGB should be (255,255,255)
print("Pixel at (0,0) - Red: {}, Green: {}0) - Red: 254, Green: 254, Blue: 254, Blue: {}".format(r,g,b)) #Notice that we are reversing the OpenCV's BGR format to RGB while printing
````

````python
# Manipulating the pixel
# Make the white pixel as black
image[0,0] = (0,0,0) # black
(b, g, r) = image[0,0]
print("Pixel at (0,0) - Red: {}, Green: {}, Blue: {}".format(r,g,b))
````

````python
# Access a part of the image and manipulate it
corner = image[0:100, 0:100] #part of the image to manipulate
cv2_to_PIL(corner)
````

#### Notice in the above image, (0,0) pixel is in black!

````python
# Manipulate the corner to be green
image[0:100, 0:100] = (0,255,0) #[start_y:end_y, start_x:end_x]
cv2_to_PIL(image)
````

````python
# Manipulate the leg region
image[150:, 75:200] = (0,255,0)
cv2_to_PIL(image)
````

### OpenCV Drawings

````python
# Create a black canvvas
canvas = np.zeros((300,300,3), dtype='uint8')
cv2_to_PIL(canvas)
````

we are using ```dtype='unit8'``` because we have the restrict the pixel values to the range of [0,255]. This is done by using 8 bit integers 

````python
# Draw a line
line_color = (0,255,0) # Green
cv2.line(canvas, (0,0),(300,300), line_color)
cv2_to_PIL(canvas)
````

````python
red_line = (0,0,255)
cv2.line(canvas,(300,0),(0,300), red_line, 3) # 3 at the end is the thickness
cv2_to_PIL(canvas)
````

````python
# Draw rectangle
# We need (top left) and (bottom right) pixels to draw a rectangle
cv2.rectangle(canvas, (10,10), (60,60), red_line,4)
cv2_to_PIL(canvas)
````

````python
# negative value for thinkness fills the rectangle rather than just outlining it
cv2.rectangle(canvas, (100,10), (160,60), red_line,-1)
cv2_to_PIL(canvas)
````

````python
# Drawing Circles
circle_canvas = np.zeros((300,300,3), dtype='uint8')
cv2_to_PIL(circle_canvas)
````

````python
# get the radius
(x,y) = (circle_canvas.shape[1]//2, canvas.shape[0]//2)
white = (255,255,255)
for r in range(0,175,25): #iterate from 0 to 175 at a step of 25 pixels
    print(r)
    cv2.circle(circle_canvas, (x,y),r,white)

cv2_to_PIL(circle_canvas)
````

````python
abstract_canvas = np.zeros((300,300,3), dtype='uint8')
cv2_to_PIL(abstract_canvas)
````

````python
# To draw a circle we need 3 values
# 1. Radius
# 2. Color
# 3. Point: (x,y) co-ordinate where the circle will be drawn
for i in range(0,25): # draw 25 random circles
    radius = np.random.randint(5, high=200) #get a radius in the range of [5,200)
    color = np.random.randint(0, high=256, size=(3,)).tolist() #specify size to get 3 random ints rather than just 1
    point = np.random.randint(0, high=300, size=(2,))
    cv2.circle(abstract_canvas,tuple(point),radius,color,-1)
cv2_to_PIL(abstract_canvas)    
````

````python
#Create the canvas
drawing_canvas = np.zeros((300,300,3), dtype='uint8')
cv2_to_PIL(drawing_canvas)
````

````python
#Draw rectangles at every 10th pixel

for (row,y) in enumerate(range(0, len(drawing_canvas),10)):
    for (col,x) in enumerate(range(0, len(drawing_canvas),10)):
        color = (0,0,255)
        if row%2 == col%2:
            color = (0,0,0)
        cv2.rectangle(drawing_canvas, (x,y), (x+10,y+10), color, -1)
cv2_to_PIL(drawing_canvas)
````

````python
#Draw the circle in the center
point = (drawing_canvas.shape[1]//2, drawing_canvas.shape[0]//2)
radius = 50
color = (0,255,0)
cv2.circle(drawing_canvas,point,radius,color,-1)
cv2_to_PIL(drawing_canvas)
````
